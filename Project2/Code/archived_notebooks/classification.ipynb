```python
from sklearn.datasets import fetch_openml

# Fetch the MNIST dataset
mnist = fetch_openml('mnist_784', version=1, as_frame=False, parser='auto')

# Extract data (features) and target (labels)
X = mnist.data
y = mnist.target

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_standard = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_standard, y, test_size=0.2, random_state=42)

#Changing the shape of y_train so that it fits our FFNN code
y_train = y_train.reshape(-1,1)
y_train = y_train.astype('int')
y_test = y_test.reshape(-1,1)
y_test = y_test.astype('int')

from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(sparse_output=False)
y_train_onehot = enc.fit_transform(y_train.reshape(-1, 1))
y_test_onehot = enc.transform(y_test.reshape(-1, 1))  
```

```python
def Call_FFNN(nodes, eta, batches, epochs, lam, output_activation, cost_func):
    if __name__ == '__main__':
        
        b_2h_50_100n_ADAMupdated = FFNN(
            nodes,
            hidden_activation = Sigmoid(),
            output_activation = output_activation, 
            cost_func = cost_func
        )

        results = b_2h_50_100n_ADAMupdated.fit(
            X = X_train, 
            t = y_train_onehot,
            scheduler = Adam(eta, rho=0.9, rho2=0.99),
            batches = batches,
            epochs = epochs,
            lam = lam
        )
```

```python
nodes = (784,100,100,10)
eta = 0.001
batches = 1
epochs = 250
lam = 0.01
hidden_activation_list = Sigmoid()
Call_FFNN(nodes, eta, batches, epochs, lam, hidden_activation_list, cost_func=MCE_multiclass())
```
